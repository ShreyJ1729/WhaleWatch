```python
import os
import matplotlib.pyplot as plt

classes = os.listdir("data")
class_dict = {}
for animal in classes:
    class_dict[animal] = len(os.listdir(f"data/{animal}"))

class_dict = {i: class_dict[i] for i in class_dict if class_dict[i]>200}
for i in class_dict:
    print(f"{i}: {class_dict[i]}")

classes = list(class_dict.keys())
```

```python
class_dict = dict(sorted(class_dict.items(), key=lambda item: item[1]))
plt.figure(figsize=(10,10))
plt.barh(y=classes, width=class_dict.values())
plt.show()
```

```python
import numpy as np
from scipy.io import wavfile
for name,count in class_dict.items():
    print("\nSpecies : ", name)
    print("\t Number of Samples = ", count)
    print("\n\t Random Sample \n")
    dirname = os.path.join("data", name)
    files = os.listdir(dirname)
    fname = os.path.join("data", name, files[np.random.choice(len(files))])
    print("\t" + fname)
    rate, data = wavfile.read(fname)
    print ("\t Sampling (frame) rate = ", rate)
    print("\t Total samples (frames) = ", data.shape)
    _ = plt.plot(data,'-', )
    plt.show()
```
